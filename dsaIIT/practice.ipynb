# Binary Search

- sort the list
- find the middle element
- compare with the element(k) we are looking for
- if it's equal, return true
- if mid element> k, apply binary search in the left half
- if mid element <k, apply binary search in the right half


```python
def binary_search(l:list,k:int):
    if l == []:
        return False 
    sorted_l=sorted(l)   
    mid=len(l)//2
    if sorted_l[mid]==k:
        return True
    if sorted_l[mid]<k:
        return binary_search(sorted_l[mid+1:],k)
    else:
        return binary_search(sorted_l[:mid],k)
```

```python
binary_search([99,5,3,5,1,3,56,34],10)
```

```python
class DSA:
    class BinarySearch:
        def __init__(self,k:int,l:list):
            self.k=k
            self.l=sorted(l)
        def search(self,l=None):
            if l is None:
                l=self.l
            if not l:
                return False
            mid=len(l)//2
            if l[mid]==self.k:
                return True
            if l[mid]<self.k:
                return self.search(l[mid+1:])
            else:
                return self.search(l[:mid])
            
```

```python
DSA.BinarySearch(5,[8,7,5,3,2,56]).search()
```

```python
class DSA:
    @staticmethod
    def binary_search(k:int,l:list):
        if l ==[]:
            return False
        l=sorted(l)
        mid=len(l)//2
        if l[mid]==k:
            return True
        if l[mid]<k:
            return DSA.binary_search(k,l[mid+1:])
        else:
            return DSA.binary_search(k,l[:mid])
```

```python
DSA.binary_search(5,[9,7,55,2,43,5,3,1,90])
```
